```python
from collections import Counter

counter = Counter(
    comp
    for comps in ds["components_used"]
    if comps
    for comp in comps
)

# Nombre de composants uniques
print(f"Nombre de composants uniques : {len(counter)}\n")

# Top 30 fréquences
print("Top 30 composants les plus fréquents :\n")
for comp, count in counter.most_common(50):
    print(f"{comp:40s} → {count:6d}")
```
